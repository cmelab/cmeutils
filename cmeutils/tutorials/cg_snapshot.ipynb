```python
import gsd
import gsd.hoomd
from cmeutils import polymers, writers
```

```python
system = polymers.System(gsd_file="80_10mers_0.8kT.gsd",
                     gsd_frame=-1,
                     atoms_per_monomer=22) 
print(system.box)

for mol in system.molecules:
    mol.sequence="PM"
    mol.assign_types()

cg_snap = writers.write_snapshot(beads=[m for m in system.monomers()])
```
